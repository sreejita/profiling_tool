&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;String <br>

```python
def is_nan(s):
    global null_count
    if isinstance(s,float) and np.isnan(s):
        null_count = null_count+1

def getNanCount(col):
    global null_count
    null_count=0
    col.apply(lambda x: is_nan(x))
    return null_count
```

```python
def is_string(string):
    flag = 1
    if isinstance(string,float) and np.isnan(string):
        flag = 0
    else:
        try:
            float(string)
            flag = 0
        
        except ValueError:
            try:
                parse(string)
                flag = 0
            except ValueError:
                if len(string) > 50:
                    flag = 0
    return flag


def get_string_ex(col):
    global string_ex
    string_ex = []
    for val in col:
        if len(string_ex) == 101:
            break
        if is_string(val) and val not in string_ex:
            string_ex.append(val)
            
    return string_ex

# get_string_ex(df['ResultCommentText'])
                
```

```python
################# Function to check if the data type of a value is date ########################


def is_date(string):
    global true_date,false_date,nan_count_date,datetime_ex
    if isinstance(string,float) and np.isnan(string):
        nan_count_date = nan_count_date+1
        # do nothing
    else:
        try:
            float(string)
        #print (string)
            false_date=false_date+1
        except ValueError:
            try:
                parse(string)
                #print (string)
                true_date=true_date+1
                if(len(datetime_ex) < 100) and string not in datetime_ex:
                    datetime_ex.append(string)
            except ValueError:
                false_date=false_date+1
            
            
def is_date_column(col):
    global true_date,false_date,nan_count_date,datetime_ex
    true_date=0
    false_date=0
    nan_count_date = 0
    datetime_ex = []

    col.apply(lambda x: is_date(x))
    return true_date,false_date,datetime_ex
#     if true_date > 0 and false_date == 0:
#         return True
#     else:
#         return False
# true_date, false_date,nan_count_date
# is_date_column(df['ActivityStartDate'])
```

```python
################# Function to check if the data type of a value is numeric ########################


def is_number(n):
    global true_numeric,false_numeric,nan_count_numeric,numeric_ex
    if (isinstance(n,float) and np.isnan(n)):
        nan_count_numeric=nan_count_numeric+1
    else:
        try:
            float(n)   # Type-casting the string to `float`.
                       # If string is not a valid `float`, 
                       # it'll raise `ValueError` exception
            #print ("num")
            if(len(numeric_ex) < 100) and n not in numeric_ex:
                numeric_ex.append(n)
            true_numeric=true_numeric+1
        except ValueError:
            #print ("not num")
            false_numeric=false_numeric+1
            
def get_numeric_stats(col):
    global true_numeric,false_numeric,nan_count_numeric,numeric_ex
    true_numeric=0
    false_numeric=0
    nan_count_numeric=0
    numeric_ex = []

    col.apply(lambda x: is_number(x))
    return true_numeric,false_numeric,numeric_ex

# get_numeric_stats(df['NormalizedResultMeasureValue'])
# true_numeric, false_numeric,nan_count_numeric
```

```python
################# Function to check if the data type of a value is text ########################


def is_text(text):
    global true_text,false_text,nan_count_text,text_ex
    if (isinstance(text,float) and np.isnan(text)):
        nan_count_text=nan_count_text+1
    else:
        if (len(text) > 50):
            true_text=true_text+1
            if(len(text_ex) < 100) and text not in text_ex:
                text_ex.append(text)
            # return True
        else:
            false_text=false_text+1
            # return False
            
def is_text_column(col):
    global true_text,false_text,nan_count_text,text_ex
    true_text=0
    false_text=0
    nan_count_text=0
    text_ex = []

    col.apply(lambda x: is_text(x))
    return true_text,false_text,text_ex
#     if true_text>0 and false_text == 0:
#         return True
#     else:
#         return False

# is_text_column(df['ResultCommentText'])
```

```python
################# Function to check if the data type of a value is categorical ########################

def is_categorical(col):
    length = len(col.unique())
#     if length > 1 and length < 5:
#         print (col.unique())
    return ( length > 1 and length < 5)
        
# is_categorical(df['ResultCommentText'])
```

```python
def profile_cols(df):
    datatypes = {}
    total_rows = len(df)
    for column in df:
        date_count,not_date_count,numeric_count, not_numeric_count,text_count,not_text_count,string_nan_count,nan_count=0,0,0,0,0,0,0,0
        dt = {}    
        numeric_count, not_numeric_count, numeric_ex = get_numeric_stats(df[column])
        if numeric_count == 0 and not_numeric_count == 0:
            #print(" none:" + column)
            dt["type"] = "None" 
        elif not_numeric_count == 0:
            dt["type"] = "Numeric"
        else:
            date_count,not_date_count, datetime_ex= is_date_column(df[column])
            if date_count > 0 and not_date_count == 0:
                dt["type"] = "DateTime" 
            else:
                text_count,not_text_count, text_ex = is_text_column(df[column])
                if text_count > 0 and not_text_count == 0:
                    dt["type"] = "Text"
                else:
                    if is_categorical(df[column]) :
                        dt["type"] = "Categorical"
                    else:
                        if date_count>0 or numeric_count >0 or text_count>0:
                            dt["type"] = "Mixed"
                            string_nan_count = total_rows - (date_count+numeric_count+text_count)
                            nan_count = getNanCount(df[column])
                            string_count = string_nan_count - nan_count
                            string_ex = get_string_ex(df[column])
                        
                            mixed_dt = {}
                            datetime_details = {}
                            numeric_details = {}
                            text_details = {}
                            string_details = {}
                            null_details = {}

                            datetime_details["count"] = date_count
                            datetime_details["examples"] = datetime_ex
                            numeric_details["count"] = numeric_count
                            numeric_details["examples"] = numeric_ex
                            text_details["count"] = text_count
                            text_details["examples"] = text_ex
                            string_details["count"] = string_count
                            string_details["examples"] = string_ex
                            null_details["count"] = nan_count

                            mixed_dt["datetime"] = datetime_details
                            mixed_dt["numeric"] = numeric_details
                            mixed_dt["text"] = text_details
                            mixed_dt["string"] = string_details
                            mixed_dt["null"] = null_details
                            
                            dt["details"] = mixed_dt
                        else:
                            dt["type"] = "String"
        datatypes[column] = dt
    return (datatypes)
```

```python
def profile_single_col(df,attr):
    datatype = ""
    total_rows = len(df)
    date_count,not_date_count,numeric_count, not_numeric_count,text_count,not_text_count,\
    string_nan_count,nan_count=0,0,0,0,0,0,0,0
   
    numeric_count, not_numeric_count, numeric_ex = get_numeric_stats(df[attr])
    if numeric_count == 0 and not_numeric_count == 0:
        datatype = "None" 
    elif not_numeric_count == 0:
        datatype = "Numeric"
    else:
        date_count,not_date_count,datetime_ex = is_date_column(df[attr])
        if date_count > 0 and not_date_count == 0:
            datatype = "DateTime" 
        else:
            text_count,not_text_count,text_ex = is_text_column(df[attr])
            if text_count > 0 and not_text_count == 0:
                datatype = "Text"
            else:
                if is_categorical(df[attr]) :
                    datatype = "Categorical"
                else:
                    if date_count>0 or numeric_count >0 or text_count>0:
                        datatype = "Mixed"
                       
                        string_nan_count = total_rows - (date_count+numeric_count+text_count)
                        
                        nan_count = getNanCount(df[attr])
                        string_count = string_nan_count - nan_count
                        string_ex = get_string_ex(df[attr])
                        
                        mixed_dt = {}
                        datetime_details = {}
                        numeric_details = {}
                        text_details = {}
                        string_details = {}
                        null_details = {}
                        
                        datetime_details["count"] = date_count
                        datetime_details["examples"] = datetime_ex
                        numeric_details["count"] = numeric_count
                        numeric_details["examples"] = numeric_ex
                        text_details["count"] = text_count
                        text_details["examples"] = text_ex
                        string_details["count"] = string_count
                        string_details["examples"] = string_ex
                        null_details["count"] = nan_count
                        
                        mixed_dt["datetime"] = datetime_details
                        mixed_dt["numeric"] = numeric_details
                        mixed_dt["text"] = text_details
                        mixed_dt["string"] = string_details
                        mixed_dt["null"] = null_details
                        
                    else:
                        datatype = "String"
    return (datatype,mixed_dt)
```

```python
def profile_col_type(df, attr, flag = 1):
    dt, mixed_dt = profile_single_col(df,attr)
    if flag == 1:
        return dt, mixed_dt
    elif flag == 0:
        return dt
    else:
        return "Flag value should be 0 or 1"
    
```

```python
def print_examples(ex, n, flag = 1):
    count = 0
    for i in range(len(ex)):
        if count == n:
            break
        if flag:
            print (ex[i])
        else:
            print("\t" + str(ex[i]))
            
        count += 1
        
    print()
```

```python
def print_col_profile_summary(df, attr, flag = 1, max_ex = 100):
    dt, mixed_dt = profile_single_col(df,attr)
    print ("Attribute: " + attr)
    print ("Datatype: " + dt)
    print()
    if flag == 1:
        if dt == "Mixed":
            if max_ex < 1 or max_ex > 100:
                print("max_ex should be between 1 and 100")
            else:
                print ("Distribution of datatypes:")
                print()
                if mixed_dt["datetime"]["count"] != 0:
                    print ("Datetime")
                    print ("Count:" + str(mixed_dt["datetime"]["count"]))
                    print ("Examples:")
                    print_examples(mixed_dt["datetime"]["examples"], max_ex)
                    print()
                
                if mixed_dt["numeric"]["count"] != 0:
                    print ("Numeric")
                    print ("Count:" + str(mixed_dt["numeric"]["count"] ))
                    print ("Examples:")
                    print_examples(mixed_dt["numeric"]["examples"], max_ex)
                    print()
                
                if mixed_dt["text"]["count"] != 0:
                    print ("Text")
                    print ("Count:" + str(mixed_dt["text"]["count"] ))
                    print ("Examples:")
                    print_examples(mixed_dt["text"]["examples"], max_ex)
                    print()
                
                if mixed_dt["string"]["count"] != 0:
                    print ("String")
                    print ("Count:" + str(mixed_dt["string"]["count"] ))
                    print ("Examples:")
                    print_examples(mixed_dt["string"]["examples"], max_ex)
                    print()
                
                if mixed_dt["null"]["count"] != 0:
                    print ("None")
                    print ("Count:" + str(mixed_dt["null"]["count"] ))
                    print()
                           
    else:
        if flag!=0:
            print ("Flag value should be 0 or 1")
    
```
